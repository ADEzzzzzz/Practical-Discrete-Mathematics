### The `sort()` function

Write a python code that count the number of character in your English Full name

```python
# Replace "Your Full Name" with your actual English full name
full_name = "Bryan Ang"

# Count the number of characters in the full name (including spaces)
character_count = len(full_name)

# Print the result
print(f"The number of characters in your full name is: {character_count}")
```

```python
# Replace "Your Full Name" with your actual English full name
full_name = "Bryan Ang"

# Remove spaces from the full name
name_without_spaces = full_name.replace(" ", "")

# Count the number of characters in the name without spaces
character_count_no_spaces = len(name_without_spaces)

# Print the result
print(f"The number of characters in your full name (excluding spaces) is: {character_count_no_spaces}")
```

```python
numbers = [3, 1, 4, 12, 8, 5, 2, 9]
names = ['Wyatt', 'Brandon', 'Kumar', 'Eugene', 'Elise']

# Apply the sort() function to the lists
numbers.sort()
names.sort()

# Display the output
print(numbers)
print(names)
```

### The `shuffle()` function

```python
import random

# Set a random seed so the code is reproducible
random.seed(1)

# Run the random.shuffle() function 5 times and display the outputs
for i in range(0,5):
    numbers = [3, 1, 4, 12, 8, 5, 2, 9]
    random.shuffle(numbers)
    print(numbers)
```
